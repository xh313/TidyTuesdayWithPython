```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.axes as ax
import matplotlib.pylab as pl
import math
from collections import Counter
import csv
import matplotlib as mpl
import matplotlib.style as style
```

```python
all = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2022/2022-02-22/freedom.csv', skiprows=0)
df = all

timespan = 26 # years
```

```python
status = list(df['Status'])
year = list(df['year'])

NFcount = {}
PFcount = {}
Fcount = {}

# Initialize the counts
for i in range(1995, 1995+timespan):
  NFcount[i] = 0
  PFcount[i] = 0
  Fcount[i] = 0


# Count the numbers
for i in range(len(year)):
  if status[i] == 'NF': # NF
    NFcount[year[i]] += 1
  if status[i] == 'F': # F
    Fcount[year[i]] += 1
  else:
    PFcount[year[i]] += 1

print(Fcount)
```

```python
# Styling

style.use('seaborn-talk')
style.use('ggplot')

mpl.font_manager.fontManager.addfont('/content/drive/MyDrive/Avenir.ttc')
mpl.rc('font', family='Avenir')  # Changing all runtime fonts into Avenir
```

```python
# Variables
nf = list(NFcount.values())
pf = list(PFcount.values())
f = list(Fcount.values())

time = np.linspace(1995, 2021, timespan)

plt.figure(dpi = 80, facecolor='#e5e5e5')
#plt.plot(time, f, '-')
plt.fill_between(time, pf, label='Partially Free', color='#8bb7d2')
plt.fill_between(time, f, label='Free', color='#448ab5')
plt.fill_between(time, nf, label='Not Free', color='#396581')

plt.ylabel('Number of Countries')#, fontdict={'size':10})
plt.xlabel('Years')#, fontdict={'size':10})
plt.xlim((1995,2020))
plt.ylim(bottom=0)
plt.title('Freedom in the World Over the Years')

plt.legend(bbox_to_anchor=(0.1, 0.05, 0.8, .102), loc='center',
          ncol=3, mode="expand", borderaxespad=0., fancybox=False)
plt.show()
```
